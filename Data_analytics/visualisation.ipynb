```python
import matplotlib.pyplot as plt
import numpy as np
```

# axes

```python
y = [22,3,33,33,56,47,23]
plt.plot(y)
```

```python
plt.figure(figsize=(15,4))
plt.pie(y)
```

```python
plt.figure(figsize=(15,4))
plt.pie(y)
plt.show()                    # har visualisation code mein plt.show() laga dena chahiye jisse graph ya pie chart , etc. mein faltu ki info naa aaye
```

```python
fig , ax = plt.subplots(nrows=2, ncols=1)
plt.show()
```

```python
fig , ax = plt.subplots(nrows=1, ncols=2, figsize=(15,2))
plt.show()
```

```python
t = np.arange(-3*np.pi,3*np.pi,0.2)
x = np.sin(t)
y = np.cos(t)
fig , ax = plt.subplots(nrows=1, ncols=2, figsize=(15,2))
ax[0].plot(t,x,'r')
ax[1].bar(t,y)
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x,y, linewidth=2, color='purple', marker='o')
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x,y, linewidth=2, color='purple', marker='*', markersize=10)
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x,y, linewidth=2, color='purple', marker='s', markersize=10 , markerfacecolor='red', markeredgecolor='green', linestyle='-.')
plt.show()
```

```python
x = [100,200,300,400,500,600,700,800,900,1000]
y = [23,45,1,34,65,2,2,46,34,56]
plt.plot(x,y,'go--' , linewidth=2, color='purple', marker='s', markersize=10 , markerfacecolor='red', markeredgecolor='green', linestyle='-.')
plt.show()
```

```python
plt.plot(x,y, linewidth=2, color='purple', marker='s', markersize=10 , markerfacecolor='red', markeredgecolor='green', linestyle='-.')
plt.title("Some graph showing something", fontdict={'fontsize':10, 'fontweight':'bold','color':'black'})
plt.xlabel("X axis ki values")
plt.ylabel("Y axis ki values", fontdict={'color':'red'})
plt.text(800,60,"This is serious \ntext here")
plt.annotate("Peak value", xy=(500,65), xytext=(600,40),arrowprops=dict(facecolor='black', shrink=0.005))
plt.show()
```
