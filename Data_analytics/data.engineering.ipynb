# Column related operations

```python
df.rename(mapper={"make":"brand"}, axis=1, inplace=True)                         # mapper helps to change names of columns
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.head()
```

```python
df.replace('alpha romeo','alfa romeo', inplace=True)
```

```python
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()                         # count the number of missing data in each column
```

```python
mnl = df['normalized-losses'].astype('float').mean()                        # Nan ka datatype float hota h # astype is doing datatye casting/conversion
```

```python
df['normalized-losses'].replace(np.nan, mnl , inplace=True)
df.head()
```

```python
bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan,bore, inplace=True)
df.head()
```

```python
stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, stroke , inplace=True)
df.head()
```

```python
price = df['price'].astype('float').mean()
df['price'].replace(np.nan, price , inplace=True)
df.head()
```

```python
df['num-of-doors'].value_counts()
```

```python
df.to_csv('automobile_cleaned.csv',index=False)
```
